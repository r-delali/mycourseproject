#scatter plot for dataset

```python
df_sample = df.sample(n=1601, random_state=42)

# Plot based on the sampled subset
plt.figure(figsize=(12, 8))
sns.scatterplot(x='fixed acidity', y='volatile acidity', data=df_sample, alpha=0.5)
plt.title('Scatter Plot: fixed acidity vs volatile acidity (Sampled Subset)')
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')
plt.show()
```
